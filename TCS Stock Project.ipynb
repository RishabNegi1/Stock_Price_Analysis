## Plotting for stock data of 2019-20

```python
#Plotting for Closing data in Scatter Plot
ts['Close'].loc['2019-01-01':'2021-01-01'].iplot()
```

```python
#Candle Plot which is only used for plotting stock data
ts.loc['2019-01-01':'2021-01-01'].iplot(kind='candle')
```

Dividends-a part of a companys profits that is paid to the people who own shares in it(shareholders)

```python
ts['Dividends'].loc['2019-01-01':'2021-01-01'].iplot()
```

Stock Splits-A stock split or stock divide increases the number of shares in a company.
A stock split causes a decrease of market price of individual shares, not causing a change of 
total market capitalization of the company.

```python
ts['Stock Splits'].loc['2019-01-01':'2021-01-01'].iplot(kind='scatter')
```

SMA-Simple Moving Average
A simple moving average (SMA) calculates the average of a selected range of prices, 
usually closing prices, by the number of periods in that range.
A simple moving average is a technical indicator that can aid in determining if an asset price will continue or if it will reverse a bull or bear trend.

```python
ts['Close'].loc['2019-01-01':'2021-01-01'].ta_plot(study='sma',periods=[13,21,55])
```

A 30-day moving average (MA) is a short term technical indicator of how stock prices are moving. It is merely the average of closing prices over the last 30 days.

```python
plt.figure(figsize=(12,6))
ts['Close'].loc['2019-01-01':'2021-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
ts['Close'].loc['2019-01-01':'2021-01-01'].plot(label='TCS Close')
plt.legend()
```
